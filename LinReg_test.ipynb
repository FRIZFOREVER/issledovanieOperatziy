```python
import pandas as pd
import numpy as np
from LinearRegression import LinearRegression
import random as rd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
np.random.seed(42)
x = np.array([np.arange(-4, 4, 0.04)])
f = lambda l: 1.2 * (l) + rd.random()*10-5 # случайное заполнение для поиска функции f = 1.2 * x + 0
y = f(x)
for i in range(200):
    y[0][i] += np.random.rand()*5
X = np.append(x, y, axis=0) # вся выборка
df = pd.DataFrame(X.transpose(), columns=['x1', 'x2'])
print(df)
sns.scatterplot(df, x='x1', y='x2');
```

```python
x = x.reshape(-1, 1)
x.shape
```

```python
y = y.reshape(-1, 1)
y.shape
```

```python
linleg = LinearRegression(
    learning_rate=0.1,
    n_iterations=100,
)
linleg.fit(x, y)
predictions_noreg = linleg.predict(x)
```

```python
linleg = LinearRegression(
    learning_rate=0.1,
    n_iterations=100,
    regularisation='L1'
)
linleg.fit(x, y)
predictions_l1 = linleg.predict(x)
```

```python
linleg = LinearRegression(
    learning_rate=0.1,
    n_iterations=100,
    regularisation='L2'
)
linleg.fit(x, y)
predictions_l2 = linleg.predict(x)
```

```python
plt.scatter(x=x, y=y, color='blue', s=6);
plt.scatter(x=x, y=predictions_noreg, color='red', s=7);
plt.scatter(x=x, y=predictions_l1, color='green', s=7);
plt.scatter(x=x, y=predictions_l2, color='pink', s=7);
```
